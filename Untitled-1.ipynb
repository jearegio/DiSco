```python
import pandas as pd
from matplotlib import pyplot as plt
import torch
import imageio
```

```python
means = torch.tensor([torch.load(f'outputs/disco/04-13-2024_19-16-12/scores_{i}.pt').mean() for i in range(1, 25)])
stds = torch.tensor([torch.load(f'outputs/disco/04-13-2024_19-16-12/scores_{i}.pt').std() for i in range(1, 25)])
```

```python
for i in range(1, 25):
    scores = torch.load(f'outputs/disco/04-13-2024_19-16-12/scores_{i}.pt')
    plt.hist(scores, range=(-900, -300), bins=50)
    plt.ylim(0, 100)
    plt.xlabel('Energy')
    plt.ylabel('Frequency')
    plt.title(f'DiSco cycle = {i}')
    plt.savefig(f'gif/scores_{i}.png')
    plt.close()
```

```python
images = []
for i in range(1, 25):
    filename = f'gif/scores_{i}.png'
    images.append(imageio.imread(filename))
imageio.mimsave('gif/movie.gif', images)
```

```python
plt.plot(torch.arange(len(means)), means, color='#B5454C')
plt.xlabel('DiSco cycles')
plt.ylabel('Energy')
plt.fill_between(torch.arange(len(means)), means-stds, means+stds, color='#FFBCBC')
```
